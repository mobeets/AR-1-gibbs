```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, invgamma
```

```python
class Params:
    def __init__(self):
        pass
```

## Data

```python
D = Params()
D.n = 1000
D.alpha = 5.0
D.beta = 0.5

D.Vv = 4.0
D.Wv = 6.0

D.V = np.random.normal(0, np.sqrt(D.Vv), D.n)
D.W = np.random.normal(D.alpha, np.sqrt(D.Wv), D.n)
D.X = np.zeros(D.n+1)
for i in xrange(D.n):
    D.X[i+1] = D.beta*D.X[i] + D.W[i]
D.Y = D.X[1:] + D.V
D.X = D.X[1:]
```

```python
plt.plot(D.X, D.Y, 'k.', lw=2, alpha=0.6)
plt.xlabel('X(t)')
plt.ylabel('Y(t)')
plt.show()

plt.plot(D.X, 'k-', lw=2, alpha=0.6)
plt.xlabel('t')
plt.ylabel('X(t)')
plt.show()

plt.plot(D.Y, 'k-', lw=2, alpha=0.6)
plt.xlabel('t')
plt.ylabel('Y(t)')
plt.show()
```

## Priors

```python
A = Params()
A.prior = Params()
B = Params()
B.prior = Params()
S = Params()
S.prior = Params()
T = Params()
T.prior = Params()

(A.prior.m, A.prior.var) = (0.0, 10.0)
(B.prior.m, B.prior.var) = (0.0, 10.0)
(S.prior.p1, S.prior.p2) = (3.0, 1.0)
(T.prior.p1, T.prior.p2) = (3.0, 1.0)

A.prior.rv = norm(loc=A.prior.m, scale=np.sqrt(A.prior.var))
B.prior.rv = norm(loc=B.prior.m, scale=np.sqrt(A.prior.var))
S.prior.rv = invgamma(S.prior.p1/2, loc=0, scale=S.prior.p1*S.prior.p2/2)
T.prior.rv = invgamma(T.prior.p1/2, loc=0, scale=T.prior.p1*T.prior.p2/2)
```

```python
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
plt.plot(x, A.prior.rv.pdf(x), 'b-', lw=5, alpha=0.6)
plt.title('alpha prior')
plt.show()

plt.plot(x, B.prior.rv.pdf(x), 'b-', lw=5, alpha=0.6)
plt.title('beta prior')
plt.show()

x = np.linspace(invgamma.ppf(0.01, S.prior.p1/2), invgamma.ppf(0.99, S.prior.p1/2), 100)
plt.plot(x, S.prior.rv.pdf(x), 'b-', lw=5, alpha=0.6)
plt.title('sig2 prior')
plt.show()

x = np.linspace(invgamma.ppf(0.01, T.prior.p1/2), invgamma.ppf(0.99, T.prior.p1/2), 100)
plt.plot(x, T.prior.rv.pdf(x), 'b-', lw=5, alpha=0.6)
plt.title('omg2 prior')
plt.show()
```

## Full conditionals

```python
A.cond = Params()
B.cond = Params()
S.cond = Params()
X1 = D.X[:-1]
X2 = D.X[1:]

"""
t := None
X := X1
Y := X2

When you want P(. | ...,X,Y)
    * you can regress X[1:] on X[:-1] to sample (a,b,s)?
    * you can take Y-X as estimates of t?
        i.e. N(0, sum((Y-X)**2)) or something like that
"""

A.cond.var = lambda (a, b, s, t, X, Y): 1/(1/A.prior.var + D.n/S.prior.p2)
A.cond.m = lambda (a, b, s, t, X, Y): A.cond.var((a,b,s,t,X,Y))*(sum(Y-b*X)/S.prior.p2 + A.prior.m/A.prior.var)
B.cond.var = lambda (a, b, s, t, X, Y): 1/(1/B.prior.var + sum(X**2)/S.prior.p2)
B.cond.m = lambda (a, b, s, t, X, Y): B.cond.var((a,b,s,t,X,Y))*(sum((Y-a)*X)/S.prior.p2 + B.prior.m/B.prior.var)
S.cond.a = lambda (a, b, s, t, X, Y): S.prior.p1 + D.n
S.cond.b = lambda (a, b, s, t, X, Y): S.prior.p1*S.prior.p2 + sum((Y-a-b*X)**2)
```

```python
A.cond.rv = lambda theta: norm(loc=A.cond.m(theta), scale=np.sqrt(A.cond.var(theta)))
B.cond.rv = lambda theta: norm(loc=B.cond.m(theta), scale=np.sqrt(B.cond.var(theta)))
S.cond.rv = lambda theta: invgamma(S.cond.p1(theta)/2, loc=0, scale=S.cond.p2(theta)/2)
```
