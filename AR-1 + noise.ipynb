```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, invgamma
```

```python
class Params:
    pass
```

## Data

```python
D = Params()
D.n = 1000
D.alpha = 10.0
D.beta = 0.5

D.Vv = 2.0
D.Wv = 6.0

D.V = np.random.normal(0, np.sqrt(D.Vv), D.n)
D.W = np.random.normal(D.alpha, np.sqrt(D.Wv), D.n)
D.X = np.zeros(D.n+1)
for i in xrange(D.n):
    D.X[i+1] = D.beta*D.X[i] + D.W[i]
D.Y = D.X[1:] + D.V
D.X = D.X[1:]
```

```python
plt.plot(D.X, D.Y, 'k.', lw=2, alpha=0.6)
plt.xlabel('X(t)')
plt.ylabel('Y(t)')
plt.show()

plt.plot(D.X, 'k-', lw=2, alpha=0.6)
plt.xlabel('t')
plt.ylabel('X(t)')
plt.show()

plt.plot(D.Y, 'k-', lw=2, alpha=0.6)
plt.xlabel('t')
plt.ylabel('Y(t)')
plt.show()
```

## Priors

```python
A = Params()
A.prior = Params()
B = Params()
B.prior = Params()
S = Params()
S.prior = Params()
T = Params()
T.prior = Params()

(A.prior.m, A.prior.var) = (10.0, 100.0)
(B.prior.m, B.prior.var) = (1.0, 2.0)
(S.prior.p1, S.prior.p2) = (8.0, 3.0)
(T.prior.p1, T.prior.p2) = (5.0, 1.0)

A.prior.rv = norm(loc=A.prior.m, scale=np.sqrt(A.prior.var))
B.prior.rv = norm(loc=B.prior.m, scale=np.sqrt(B.prior.var))
S.prior.rv = invgamma(S.prior.p1/2, loc=0, scale=S.prior.p1*S.prior.p2/2)
T.prior.rv = invgamma(T.prior.p1/2, loc=0, scale=T.prior.p1*T.prior.p2/2)
```

```python
plt.subplot(221)
x = np.linspace(A.prior.rv.ppf(0.001), A.prior.rv.ppf(0.999), 100)
plt.plot(x, A.prior.rv.pdf(x), 'b-', lw=5, alpha=0.6)
plt.title('alpha prior')

plt.subplot(222)
x = np.linspace(B.prior.rv.ppf(0.001), B.prior.rv.ppf(0.999), 100)
plt.plot(x, B.prior.rv.pdf(x), 'b-', lw=5, alpha=0.6)
plt.title('beta prior')

plt.subplot(223)
x = np.linspace(S.prior.rv.ppf(0.01), S.prior.rv.ppf(0.99), 100)
plt.plot(x, S.prior.rv.pdf(x), 'b-', lw=5, alpha=0.6)
plt.title('sig2 prior')

plt.subplot(224)
x = np.linspace(T.prior.rv.ppf(0.01), T.prior.rv.ppf(0.99), 100)
plt.plot(x, T.prior.rv.pdf(x), 'b-', lw=5, alpha=0.6)
plt.title('ome2 prior')

plt.tight_layout()
plt.show()
```

## Full conditionals

```python
A.cond = Params()
B.cond = Params()
S.cond = Params()
T.cond = Params()

# AR-1 without noise; regress X[1:] on X[:-1]
A.cond.var = lambda (a, b, s, t, X, Y): 1/(1/A.prior.var + D.n/S.prior.p2)
A.cond.m = lambda (a, b, s, t, X, Y): A.cond.var((a,b,s,t,X,Y))*(sum(X[1:]-b*X[:-1])/S.prior.p2 + A.prior.m/A.prior.var)
B.cond.var = lambda (a, b, s, t, X, Y): 1/(1/B.prior.var + sum(X**2)/S.prior.p2)
B.cond.m = lambda (a, b, s, t, X, Y): B.cond.var((a,b,s,t,X,Y))*(sum((X[1:]-a)*X[:-1])/S.prior.p2 + B.prior.m/B.prior.var)
S.cond.p1 = lambda (a, b, s, t, X, Y): S.prior.p1 + D.n
S.cond.p2 = lambda (a, b, s, t, X, Y): S.prior.p1*S.prior.p2 + sum((X[1:]-a-b*X[:-1])**2)

# Regression of Y on X
T.cond.p1 = lambda (a, b, s, t, X, Y): T.prior.p1 + D.n
T.cond.p2 = lambda (a, b, s, t, X, Y): T.prior.p1*T.prior.p2 + sum((Y-X)**2)
```

```python
A.cond.rv = lambda theta: norm(loc=A.cond.m(theta), scale=np.sqrt(A.cond.var(theta)))
B.cond.rv = lambda theta: norm(loc=B.cond.m(theta), scale=np.sqrt(B.cond.var(theta)))
S.cond.rv = lambda theta: invgamma(S.cond.p1(theta)/2, loc=0, scale=S.cond.p1(theta)/2)
T.cond.rv = lambda theta: invgamma(T.cond.p1(theta)/2, loc=0, scale=T.cond.p1(theta)/2)

def draw_theta((aCur, bCur, sCur, tCur), X, Y):
    aCur = A.cond.rv((aCur, bCur, sCur, tCur, X, Y)).rvs(size=1)
    bCur = B.cond.rv((aCur, bCur, sCur, tCur, X, Y)).rvs(size=1)
    sCur = S.cond.rv((aCur, bCur, sCur, tCur, X, Y)).rvs(size=1)
    tCur = T.cond.rv((aCur, bCur, sCur, tCur, X, Y)).rvs(size=1)
    return (aCur, bCur, sCur, tCur)
```

## Filter Forward Backwards Sampling (FFBS)

```python
def inner_backward((a,b,s,t), Xt1, Yt1, mt, Ct, rt1):
    """
    n.b. read Xt1 as "X_{t+1}"
    """
    ht = mt + (b * Ct)/(Yt1*(Xt1 - rt1))
    # ht = mt + (b * Ct * (Xt1 - rt1))/Yt1
    Ht = Ct - (b**2)*(Ct**2)/Yt1
    if Ht <= 0:
        print (a,b,s,t)
        print "Invalid Ht: {0} = {1} - {2}".format(Ht, Ct, (b**2)*(Ct**2)/Yt1)
    return norm(loc=ht, scale=np.sqrt(Ht)).rvs(size=1)

def backward((a,b,s,t), XT, Y, m, C, r):
    """
    generates each Xt, given each X_{t+1}
    """
    X = np.zeros(len(Y))
    X[-1] = XT
    for i in xrange(len(Y)-2, -1, -1):
        X[i] = inner_backward((a,b,s,t), X[i+1], Y[i+1], m[i], C[i], r[i+1])
    return X

def forward((a,b,s,t), Y):
    """
    generates the last X, and parameters for its distribution N(m,C)
    """
    m = np.zeros(len(Y))
    C = np.zeros(len(Y))
    r = np.zeros(len(Y))
    R = np.zeros(len(Y))
    f = np.zeros(len(Y))
    F = np.zeros(len(Y))
    h = np.zeros(len(Y))
    H = np.zeros(len(Y))
    for i in xrange(len(Y)):
        # prior at i-1 ~ N(r[i], R[i])
        r[i] = a + b*m[i-1]
        R[i] = (b**2)*C[i] + t
        # predictive at i-1 ~ N(f[i], F[i])
        f[i] = r[i]
        F[i] = R[i] + s
        # posterior at i ~ N(m[i], C[i])
        h[i] = Y[i] - f[i]
        H[i] = R[i]/F[i]
        m[i] = r[i] + H[i]*h[i]
        C[i] = R[i] - R[i]*H[i]
    if C[-1] <= 0:
        print "Invalid CT: {0}".format(C[-1])
    XT = norm(loc=m[-1], scale=np.sqrt(C[-1])).rvs(size=1)
    return XT, m, C, r

def ffbs(theta, Y):
    XT, m, C, r = forward(theta, Y)
    X = backward(theta, XT, Y, m, C, r)
    return X
```

## Gibbs Sampling

```python
niters = 1000
Thetas = np.zeros([niters+1, 4])
States = np.zeros([niters+1, len(D.Y)])
States[0] = D.Y
for i in xrange(1, niters+1):
    if i % 10 == 0:
        print '{0}%'.format(int(i*100.0/(niters+1))),
    Thetas[i] = draw_theta(Thetas[i-1], States[i-1], D.Y)
    States[i] = ffbs(Thetas[i], D.Y)
tMAP = Thetas.mean(0)
print
print '(alpha, beta, sig2, ome2) = {0}'.format(tMAP)
```

```python
# tMAP = [ 5.33132064, 0.47778493, 5.64343147]
alpha, beta, sig2, ome2 = tMAP
pMAP = (alpha, beta, sig2, ome2, States[-1], D.Y)

x1 = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
x2 = np.linspace(invgamma.ppf(0.01, S.prior.p1/2), invgamma.ppf(0.99, S.prior.p1/2), 100)
x3 = np.linspace(invgamma.ppf(0.01, T.prior.p1/2), invgamma.ppf(0.99, T.prior.p1/2), 100)

plt.subplot(221)
plt.plot(x1, A.cond.rv(pMAP).pdf(x1), 'r-', lw=5, alpha=0.6, label='a prior')
plt.title('alpha conditional at MAP')

plt.subplot(222)
plt.plot(x1, B.cond.rv(pMAP).pdf(x1), 'r-', lw=5, alpha=0.6, label='b prior')
plt.title('beta conditional at MAP')

plt.subplot(223)
plt.plot(x2, S.cond.rv(pMAP).pdf(x2), 'r-', lw=5, alpha=0.6, label='sig2 prior')
plt.title('sig2 conditional at MAP')

plt.subplot(224)
plt.plot(x3, T.cond.rv(pMAP).pdf(x3), 'r-', lw=5, alpha=0.6, label='sig2 prior')
plt.title('sig2 conditional at MAP')

plt.tight_layout()
plt.show()
```
