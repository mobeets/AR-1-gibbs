```python
import numpy as np
from datetime import datetime
import pandas.io.data as web
import matplotlib.pyplot as plt
```

```python
df1 = datetime(2013, 1, 1)
dfX = web.DataReader("GOOG", 'google', df1)
dfY = web.DataReader("YHOO", 'google', df1)
inds = dfX.index & dfY.index
X = dfX['Close'][inds].values
Y = dfY['Close'][inds].values
X = X - np.mean(X)
Y = Y - np.mean(Y)
```

```python
plt.subplot(411)
plt.plot(X, Y, 'k.', lw=2, alpha=0.6)
plt.subplot(412)
plt.plot(X[:-1], X[1:], 'k.', lw=2, alpha=0.6)
plt.subplot(413)
plt.plot(X, 'k.', lw=2, alpha=0.6)
plt.subplot(414)
plt.plot(Y, 'k.', lw=2, alpha=0.6)
plt.tight_layout()
```

```python
def regress(Y, X, delta=0.4):
    N = len(X)
    m = np.zeros(N+1)
    C =np.zeros(N+1)
    R = np.zeros(N+1)
    f = np.zeros(N+1)
    Q = np.zeros(N+1)
    S = np.zeros(N+1)
    e = np.zeros(N+1)
    A = np.zeros(N+1)
    n = np.zeros(N+1)
    d = np.zeros(N+1)
    
    m[0] = 0.1
    C[0] = 1.0
    n[0] = 1.0
    d[0] = 1.0
    S[0] = d[0]/n[0]
    X = np.hstack([[0], X])
    Y = np.hstack([[0], Y])
    for t in xrange(1,N+1):
        # prior at t
        R[t] = C[t-1]/delta
        f[t] = m[t-1]*X[t]
        Q[t] = (X[t]**2)*R[t] + S[t-1]
#         print (R[t], f[t], Q[t])
        # posterior at t
        e[t] = Y[t] - f[t]
        A[t] = X[t]*R[t]/Q[t]
        n[t] = n[t-1] + 1
        d[t] = d[t-1] + S[t-1]*(e[t]**2)/Q[t]
        S[t] = d[t]/n[t]
        m[t] = m[t-1] + A[t]*e[t]
        C[t] = R[t]*S[t]/Q[t]
#         print (e[t], A[t], n[t], d[t], S[t], m[t], C[t])
    return m, S
```

```python
X = X - np.mean(X)
Y = Y - np.mean(Y)
A = np.vstack([X, np.ones(len(X))]).T
d0, c = np.linalg.lstsq(A, Y)[0]
print d0, c
for delta in np.linspace(0.01, 0.99, 3):
    mh, Sh = regress(Y, X, delta)
    plt.ylim([-0.2, 0.35])
    plt.plot([0, len(mh)], [d0, d0], '--', lw=2, alpha=0.4)
    plt.plot(mh, 'k.', lw=2, alpha=0.6)
    plt.show()
    plt.plot(Sh, 'k.', lw=2, alpha=0.6)
    plt.show()
```
